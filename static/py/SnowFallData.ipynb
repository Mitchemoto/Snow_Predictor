```python
#import dependencies
import pandas as pd
import numpy as np
import glob
import os
```

```python
#bring in file path
path = r'../Data'
#use glob to read through all files
all_files=glob.glob(path + "/*.csv")
#create empty list to fill with csv data
li =[]
#create for loop to reach through each file and append to li
for filename in all_files:
    snowDataDf=pd.read_csv(filename, index_col=None, header=0)
    li.append(snowDataDf)

snowframe = pd.concat(li, axis=0, ignore_index=True)
snowframe.head()
```

```python
cleanSnowDf=snowframe.drop(['Station Id','Unnamed: 12'],axis=1)
cleanSnowDf.head()
```

```python
len(cleanSnowDf.index)
```

```python
orderedSnowDf=cleanSnowDf[['Date','Station Name','Snow Water Equivalent (in) Start of Day Values','Change In Snow Water Equivalent (in)',
                                            'Snow Depth (in) Start of Day Values','Change In Snow Depth (in)','Elevation (ft)','Air Temperature Average (degF)',
                                            'Latitude','Longitude','County Name']]
orderedSnowDf.head(15)
```

```python
orderedSnowDf.to_csv(r'../Data/SnowData.csv')
```
